```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Data_for_Transformation.csv')
# print(data)
```

```python
plt.scatter(data['Age'],data['Salary'])
plt.show()
```

```python
plt.hist(data['Salary'],bins=6)
plt.show()
```

```python
countrylist=data['Country']
d=dict()
print(countrylist)
for i in countrylist:
    if i not in d:
        d.update({i:1})
    else:
        d[i]+=1
print(d)        
countrycount=[]
countryname=[]
for i in d:
    countrycount.append(d[i])
    countryname.append(i)

fig=plt.figure(figsize=(5,5))
plt.pie(countrycount,labels=countryname,autopct='%.2f%%');
plt.show()
```
